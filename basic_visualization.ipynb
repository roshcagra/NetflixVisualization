```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv('project3data/data.txt', sep='\t', header=None, names = ['user','movie','rating'])
movie_labels = pandas.read_csv('project3data/movies.txt', sep='\t', header=None, index_col=0, names = ['Movie Title', 'Unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'])
```

```python
def rating_histrogram(data, title_label):
    rating_plot = data.rating.value_counts().sort_index().plot(kind='bar', title='Distribution of Ratings for ' + title_label)
    rating_plot.set_xlabel('Rating')
    rating_plot.set_ylabel('Frequency')
    plt.show()
```

```python
#1. All Data
rating_histrogram(data, 'All Movies')
```

```python
#2. 10 Most Popular Movies
ten_most_popular_movies = data.movie.value_counts().index[0:10]
ratings_for_ten_most_popular = data[data['movie'].isin(ten_most_popular_movies)]
rating_histrogram(ratings_for_ten_most_popular, '10 Most Popular Movies')
```

```python
#3. 10 Highest Rated Movies
ten_highest_rated_movies = data.groupby('movie')['rating'].mean().sort_values(ascending=False).index[0:10]
ratings_for_ten_highest_rated = data[data['movie'].isin(ten_highest_rated_movies)]
rating_histrogram(ratings_for_ten_highest_rated, '10 Highest Rated Movies')
```

```python
#4. Three Categories
comedies = movie_labels[movie_labels['Comedy'] == 1].index
ratings_for_comedies = data[data['movie'].isin(comedies)]
rating_histrogram(ratings_for_comedies, 'Comedies')

documentaries = movie_labels[movie_labels['Documentary'] == 1].index
ratings_for_documentaries = data[data['movie'].isin(documentaries)]
rating_histrogram(ratings_for_documentaries, 'Documentaries')

film_noir_movies = movie_labels[movie_labels['Film-Noir'] == 1].index
ratings_for_film_noir_movies = data[data['movie'].isin(film_noir_movies)]
rating_histrogram(ratings_for_film_noir_movies, 'Film-Noir Movies')
```
